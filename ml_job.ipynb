```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import csv 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\user\Documents\ml_python\Salary\Salary.csv")
```

```python
print(df)
```

```python
df.plot(x='YearsExperience', y='Salary', kind='line', marker='o', linestyle='-', color='b')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Relation between Experience and salary')
```

```python
experience_train, experience_test, salary_train, salary_test = train_test_split(df.YearsExperience, df.Salary, test_size=0.2, random_state=42)
```

```python
#print(experience_train)
```

```python
model = LinearRegression()
```

```python
experience_train=experience_train.values.reshape(-1, 1)
```

```python
model.fit(experience_train, salary_train)
```

```python
experience_test=experience_test.values.reshape(-1, 1)
```

```python
salary_pred = model.predict(experience_test)
```

```python
print(salary_pred)
```

```python
plt.scatter(experience_test, salary_test, label='Actual Salary', color='red')
plt.plot(experience_test, salary_pred, label='Regression Line', color='blue', linewidth=3)
plt.xlabel('Experience')
plt.ylabel('Salary ')
plt.legend()
```
